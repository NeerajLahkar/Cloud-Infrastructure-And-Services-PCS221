Q1. Create 20000 × 10000 data file using random module

```python
import random
import csv

rows = 20000
cols = 10000

with open("large_matrix.csv", "w", newline="") as file:
    writer = csv.writer(file)

    for _ in range(rows):
        row = random.choices(range(100), k=cols)
        writer.writerow(row)

```

Q2. Assign subsequent rows to matrix first row elements

```python
import numpy as np

# Example matrix
matrix = np.array([
    [10, 20, 30],
    [1, 2, 3],
    [4, 5, 6]
])

first_row = matrix[0]
remaining_rows = matrix[1:]

result = {}

for i in range(len(first_row)):
    result[first_row[i]] = remaining_rows[:, i]

print(result)
```

Q3. Group similar elements into matrix

```python
import numpy as np
from collections import defaultdict

matrix = np.array([
    [1, 2, 2],
    [3, 1, 3],
    [2, 3, 1]
])

groups = defaultdict(list)

for i in range(matrix.shape[0]):
    for j in range(matrix.shape[1]):
        groups[matrix[i][j]].append((i, j))

print("Grouped Elements:")
for key, value in groups.items():
    print(key, ":", value)
```

Q4. Add and Subtract Matrices

```python
import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

addition = A + B
subtraction = A - B

print("Addition:\n", addition)
print("Subtraction:\n", subtraction)
```

Q5. Create n × n matrix where opposite corners sum is even

```python
import numpy as np

n = int(input("Enter n: "))

matrix = np.zeros((n, n), dtype=int)

for i in range(n):
    for j in range(n):
        matrix[i][j] = 2 * (i + j)  # ensures even numbers

print("Matrix:\n", matrix)
```

Q6. Row-wise element addition in tuple matrix

```python
matrix = (
    (1, 2, 3),
    (4, 5, 6),
    (7, 8, 9)
)

row_sums = []

for row in matrix:
    row_sums.append(sum(row))

print("Row-wise sums:", row_sums)
```
