# QUESTION 1:

(A) Manual Python Approach

```python
data = ["hadoop is fast", "hadoop is scalable"]

mapped = []
for line in data:
    words = line.split()
    for word in words:
        mapped.append((word, 1))

print("Mapper Output:", mapped)
```

```python
from collections import defaultdict

shuffled = defaultdict(list)

for word, count in mapped:
    shuffled[word].append(count)

print("Shuffled Output:", dict(shuffled))
```

```python
reduced = {}

for word, counts in shuffled.items():
    reduced[word] = sum(counts)

print("Final Word Count:", reduced)
```

```python
# Step 1: Write mrjob script to file
with open("wordcount.py", "w") as f:
    f.write("""
from mrjob.job import MRJob

class WordCount(MRJob):

    def mapper(self, _, line):
        for word in line.split():
            yield word, 1

    def reducer(self, key, values):
        yield key, sum(values)

if __name__ == "__main__":
    WordCount.run()
""")

# Step 2: Create input file
with open("input.txt", "w") as f:
    f.write("""hadoop is fast
hadoop is scalable""")

# Step 3: Run mrjob
!python wordcount.py input.txt
```

```python
# Step 1: Write mrjob script
with open("avg_word_length.py", "w") as f:
    f.write("""
from mrjob.job import MRJob

class AvgWordLength(MRJob):

    def mapper(self, _, line):
        for word in line.split():
            yield word, (len(word), 1)

    def reducer(self, key, values):
        total_length = 0
        count = 0
        for length, c in values:
            total_length += length
            count += c
        yield key, total_length / count

if __name__ == "__main__":
    AvgWordLength.run()
""")

# Step 2: Create input file
with open("input.txt", "w") as f:
    f.write("data science data big data")

# Step 3: Run job
!python avg_word_length.py input.txt
```

# QUESTION 4:

(A) Manual Python Approach

```python
data = ["hadoop mapreduce spark"]

mapped = []
for line in data:
    for word in line.split():
        mapped.append(("global", (len(word), 1)))

print("Mapper Output:", mapped)
```

```python
from collections import defaultdict

shuffled = defaultdict(list)

for key, value in mapped:
    shuffled[key].append(value)

print("Shuffled Output:", dict(shuffled))
```

```python
reduced = {}

for key, values in shuffled.items():
    total_length = sum(v[0] for v in values)
    total_count = sum(v[1] for v in values)
    reduced[key] = total_length / total_count

print("Final Output:", reduced)
```

# QUESTION 5:

(A) Manual Python Approach

```python
import re
from collections import Counter, defaultdict

# Step 1: Read file
with open("shakespeare.txt", "r") as f:
    text = f.read().lower()

# Step 2: Clean text (VERY IMPORTANT for real dataset)
words = re.findall(r'\b[a-z]+\b', text)

# -------------------------
# Q1: Word Count
# -------------------------
word_count = Counter(words)
print("Top 10 Word Count:", word_count.most_common(10))


# -------------------------
# Q3: Avg Word Length (per word)
# -------------------------
d = defaultdict(list)

for word in words:
    d[word].append(len(word))

avg_word_len = {k: sum(v)/len(v) for k,v in d.items()}

print("\nSample Avg Word Length:")
for i, (k, v) in enumerate(avg_word_len.items()):
    if i == 5:
        break
    print(k, ":", v)


# -------------------------
# Q4: Global Average
# -------------------------
total_len = sum(len(w) for w in words)
total_words = len(words)

print("\nGlobal Avg Word Length:", total_len / total_words)


# -------------------------
# ⭐ Top 5 Words
# -------------------------
print("\nTop 5 Words:", word_count.most_common(5))
```

(B) mrjob Framework

```python
# Step 1: Write mrjob script
with open("top5_shakespeare.py", "w") as f:
    f.write("""
from mrjob.job import MRJob
import re
from collections import Counter

class Top5Words(MRJob):

    def mapper(self, _, line):
        words = re.findall(r'\\b[a-z]+\\b', line.lower())
        for word in words:
            yield "all", word

    def reducer(self, key, words):
        c = Counter(words)
        for word, count in c.most_common(5):
            yield word, count

if __name__ == "__main__":
    Top5Words.run()
""")

# Step 2: Run job
!python top5_shakespeare.py shakespeare.txt
```

# QUESTION 7:

(A) Manual Python Approach

```python
data = ["HR 50000", "IT 70000", "HR 60000", "IT 80000"]

mapped = []
for line in data:
    dept, salary = line.split()
    mapped.append((dept, int(salary)))

print("Mapper Output:", mapped)
```

```python
from collections import defaultdict

shuffled = defaultdict(list)

for dept, salary in mapped:
    shuffled[dept].append(salary)

print("Shuffled Output:", dict(shuffled))
```

```python
avg_salary = {}

for dept, salaries in shuffled.items():
    avg_salary[dept] = sum(salaries) / len(salaries)

print("Average Salary:", avg_salary)
```

```python
highest_dept = max(avg_salary, key=avg_salary.get)

print("Highest Paid Department:", highest_dept)
```

(B) mrjob Framework

```python
# Step 1: Write mrjob script
with open("avg_salary.py", "w") as f:
    f.write("""
from mrjob.job import MRJob

class AvgSalary(MRJob):

    def mapper(self, _, line):
        dept, salary = line.split()
        yield dept, int(salary)

    def reducer(self, key, values):
        total = 0
        count = 0
        for v in values:
            total += v
            count += 1
        yield key, total / count

if __name__ == "__main__":
    AvgSalary.run()
""")

# Step 2: Create input file
with open("input.txt", "w") as f:
    f.write("""HR 50000
IT 70000
HR 60000
IT 80000""")

# Step 3: Run mrjob
output = !python avg_salary.py input.txt

# Step 4: Parse output safely
avg_dict = {}

for line in output:
    if line.startswith('"'):   # only actual results
        parts = line.strip().split("\t")   # tab-separated
        dept = parts[0].strip('"')
        value = float(parts[1])
        avg_dict[dept] = value

# Step 5: Find highest paid department
highest = max(avg_dict, key=avg_dict.get)

# Final prints
print("Averages:", avg_dict)
print("Highest Paid Department:", highest)
```

# QUESTION 9:

(A) Manual Approach

```python
import csv
from collections import defaultdict

# Dictionary to store temps
d = defaultdict(list)

# Read file
with open("GlobalLandTemperatures_GlobalLandTemperaturesByMajorCity.csv", "r") as f:
    reader = csv.DictReader(f)

    for row in reader:
        try:
            city = row["City"]
            temp = float(row["AverageTemperature"])

            # Ignore missing values
            if temp:
                d[city].append(temp)
        except:
            continue

# Compute average
avg_temp = {k: sum(v)/len(v) for k, v in d.items()}

# Show sample output (first 5)
print("Sample Output:")
for i, (k, v) in enumerate(avg_temp.items()):
    if i == 5:
        break
    print(k, ":", v)
```

(B) mrjob Framework

```python
# Step 1: Write mrjob script
with open("avg_temp_city.py", "w") as f:
    f.write("""
from mrjob.job import MRJob
import csv

class AvgTempCity(MRJob):

    def mapper(self, _, line):
        row = line.split(",")

        try:
            city = row[3]
            temp = float(row[1])

            if temp:
                yield city, temp
        except:
            pass

    def reducer(self, key, values):
        total = 0
        count = 0
        for v in values:
            total += v
            count += 1
        yield key, total / count

if __name__ == "__main__":
    AvgTempCity.run()
""")

# Step 2: Run mrjob
output = !python avg_temp_city.py GlobalLandTemperatures_GlobalLandTemperaturesByMajorCity.csv

# Step 3: Parse safely
avg_dict = {}

for line in output:
    if line.startswith('"'):
        parts = line.strip().split("\t")
        city = parts[0].strip('"')
        value = float(parts[1])
        avg_dict[city] = value

# Show sample output
print("Sample Output:")
count = 0
for k, v in avg_dict.items():
    print(k, ":", v)
    count += 1
    if count == 5:
        break
```
